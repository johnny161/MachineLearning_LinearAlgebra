## 复数和复平面
欧拉公式：$e^{i\theta}=cos\theta+isin\theta$  
对于这种模长（半径）为1的复数，它具备一个很好的性质：  
它的n次幂运算结果所得的复数半径始终为1，保持不变，只有角$\theta$变为原来的n倍，也就是说幂运算的结果始终在复平面的单位圆上滑动

```python
z1 = 2 + 1j
z2 = 3 + 5j
print(z1.real) #实部
print(z2.imag) #虚部
print(z1+z2) #加法
print(z1*z2) #乘法
print(z1.conjugate()) #共轭
print(abs(z1)) #模长
```

厄米矩阵S的特征值一定是实数，不同特征值对应的特征向量满足彼此正交

```python
# 时域信号在一个2pi周期内的形态
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f(x):
    return np.sin(x)+2*np.sin(3*x)+2*np.cos(3*x)+4*np.sin(15*x)

x = np.linspace(0, 2*np.pi, 2048)
plt.scatter(x, f(x))
plt.grid()
plt.show()
```

```python
# 对这段时域信号进行频域分析
# 图中的3个能量最高的峰值点，正对应时域函数中的三个谐波频率
import numpy as np
from scipy.fftpack import fft
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 128)
y = np.sin(x)+2*np.sin(3*x)+2*np.cos(3*x)+4*np.sin(15*x)

xf = np.arange(len(y))             # 离散频率
xf_half = xf[range(int(len(x)/2))] # 由于对称性 因此只取一半区域
yf = abs(fft(y))/len(x)            # 执行完fft后 对各频率的能量进行归一化处理
yf_half = yf[range(int(len(x)/2))] # 由于对称性 因此只取一半区间

plt.plot(xf_half, yf_half)
plt.show()
```
